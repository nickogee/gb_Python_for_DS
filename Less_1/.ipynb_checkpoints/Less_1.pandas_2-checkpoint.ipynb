```python
import numpy as np
import pandas as pd
```

Слияние данных

```python
autors = pd.DataFrame({
    'author_id':[1,2,3],
    'author_name':['Pushkin', 'Tolstoy', 'Dostoevskiy'],
    
})
autors
```

```python
books = pd.DataFrame({
    'author_id':[2,3,3,4],
    'book_title': ['Война и мир', 'Идиот', 'Преступление и наказание', 'Отцы и дети']
})
books
```

```python
df1 = pd.merge(autors, books, on='author_id', how='inner')
df1
```

```python
df2 = pd.merge(autors, books, on='author_id', how='left')
df2
```

```python
df3 = pd.merge(autors, books, on='author_id', how='right')
df3
```

```python
df4 = pd.merge(autors, books, on='author_id', how='outer')
df4
```

Работа с пропущенными данными

```python
df4.loc[df4['book_title'].isnull(), :]
```

```python
df4.loc[df4['book_title'].notnull(), :]
```

```python
df4['book_title'] = df4['book_title'].fillna('unknown')
df4
```

```python
df4['author_name'].fillna('unknown', inplace=True) 
df4
```

Добавление столдцов

```python
df4.loc[(df4['author_name'] != 'unknown') & (df4['book_title'] != 'unknown'), 'quantity'] = 1
df4
```

```python
df4['quantity'].fillna(0, inplace=True)
df4
```

```python
df4['quantity'] = df4['quantity'].astype(int)
df4
```

```python
df4.set_index('author_id', inplace=True)
df4
```

```python
df4.reset_index(inplace=True)
df4
```

```python
df4['Price'] = 500
df4
```

```python
df4 = df4.drop('Price', axis=1)
df4
```

```python
df4 = df4.drop(1, axis=0)
df4
```

Вернем удаленную строку

```python
df4 = df4.append(pd.DataFrame({
    'author_id': [2],
    'author_name': ['Tolstoy'],
    'book_title': ['Война и Мир'],
    'quantity': [1]
}), ignore_index=True)
df4
```

Отсортитруем

```python
df4 = df4.sort_values(by='author_id')
df4
```

Переустановим индекс

```python
df4 = df4.reset_index(drop=True)
df4
```

Соединение датафреймов

```python
df5 = pd.DataFrame({
    'author_id': [3,5],
    'author_name': ['Dostoevskiy', 'Chehov'],
    'book_title': ['Игрок', 'Три сестры'],
    'quantity': [2, 3]
})
df5
```

```python
df6 = pd.concat([df4, df5], axis=0, ignore_index=True)
df6
```

```python
prices = pd.DataFrame({
    'price': [700, 450, 500, 400, 350]
}, columns=['price'], index=[1,2,3,5,6])
prices
```

```python
df7 = pd.concat([df6, prices], axis=1)
df7
```

```python
df7['total'] = df7['quantity'] * df7['price']
df7
```
