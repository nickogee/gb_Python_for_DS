```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

### Форматирование текста для графика

##### Числа от -5 до 5, 101 значений


```python
x = np.linspace(-5, 5, 101)
x
```

```python
x2 = x**2
x2
```

```python
plt.plot(x, x2)
```

```python
plt.plot(x, x2)
plt.title('График f(x)')
```

```python
# Поменяем настройки графики: применим SVG - масштабируемая векторная графика
%config InlineBackend.figure_format= 'svg'
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#004353'}
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontdict=title_dict)
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontdict=title_dict, loc='left')
```

### Добавление названий осей

```python
plt.plot(x, x2)
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х')
plt.ylabel('Функция f(x)')
```

```python
lable_dict = {'fontsize': 10, 'color': '#999999'}
```

```python
plt.plot(x, x2)
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')
```

```python
x3 = x**3
x3
```

```python
plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')
```

```python
plt.axis([-7, 7, -10, 10])

plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')
```

```python
plt.axis([-7, 7, -10, 10])

plt.plot(x, x2)
plt.plot(x, x3)
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')

plt.legend(labels=['x2', 'x3'])
```

```python
plt.axis([-7, 7, -10, 10])

plt.plot(x, x2, label='x**2', color='red')
plt.plot(x, x3, label='x**3', color='#005588')
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')

plt.legend(loc='lower right', frameon=False)
```

```python
plt.axis([-7, 7, -10, 10])

plt.plot(x, x2, label='x**2', color='red', linestyle='dotted')
plt.plot(x, x3, label='x**3', color='#005588', linestyle='dashed')
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')

plt.legend(loc='lower right', frameon=False)
```

```python
plt.axis([-7, 7, -10, 10])

plt.plot(x, x2, label='x**2', color='red', linestyle='dotted')
plt.plot(x, x3, label='x**3', color='#005588', linestyle='dashed')
plt.title('График f(x)', fontdict=title_dict)
plt.xlabel('Переменная Х', fontdict=lable_dict)
plt.ylabel('Функция f(x)')

plt.legend(loc='lower right', frameon=False)
plt.grid()
```
