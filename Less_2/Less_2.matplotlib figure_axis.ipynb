```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

### Явно зададим axes и figure

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)

fig.set_size_inches(8.5, 3,5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)

fig.set_size_inches(8.25, 3,5)
plt.subplots_adjust(wspace=0.7)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)

fig.set_size_inches(3, 6)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)

fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.7)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('x**2')

ax2.plot(x, x3)
ax2.set_title('x**3')

ax3.plot(x, y)
ax3.set_title('sin(x)')

ax4.plot(x, z)
ax4.set_title('cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('x**2')

ax2.plot(x, x3)
ax2.set_title('x**3')

ax3.plot(x, y)
ax3.set_title('sin(x)')

ax4.plot(x, z)
ax4.set_title('cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('x**2')
ax1.set_xlim([-3, 3])

ax2.plot(x, x3)
ax2.set_title('x**3')
ax2.set_xlim([-20, 20])

ax3.plot(x, y)
ax3.set_title('sin(x)')

ax4.plot(x, z)
ax4.set_title('cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(hspace=0.3)
plt.subplots_adjust(wspace=0.4)
```
