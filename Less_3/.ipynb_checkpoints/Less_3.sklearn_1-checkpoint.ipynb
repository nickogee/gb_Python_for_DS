```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
boston_url = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None, names=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
target = boston_url.values[1::2, 2]
```

```python
boston_csv = pd.read_csv('boston_house_prices.csv', header=1)
boston_csv.head()
```

### Будем использовать данные по url, предварительно их очистив

###  датасет с данными

```python
x = boston_url.copy()
x
```

```python
# только для датасета по url
x = x.loc[:,'CRIM':'PTRATIO'].dropna()
x
```

### датасет с целевыми ценами

```python
y = pd.DataFrame(target, columns=['price'])
y
```

### разбиваем данные на тренеровочные и тестовые

```python
from sklearn.model_selection import train_test_split
```

```python
x_trane, x_test, y_trane, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

### построение модели линейной регрессии

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_trane, y_trane)
```

```python
y_pred = lr.predict(x_test)
```

```python
check_test = pd.DataFrame(
        {'y_test': y_test['price'],
        'y_pred': y_pred.flatten()
        }, columns=['y_test', 'y_pred']
)
```

```python
check_test
```

### метрики оценки качества регрессионных моделей

#### средняя квадратичная ошибка

```python
check_test['error'] = check_test['y_pred']- check_test['y_test']
```

```python
check_test.head()
```

```python
initial_mse = (check_test['error'] **2).mean()
initial_mse
```

```python
from sklearn.metrics import mean_squared_error
```

```python
initial_mse = mean_squared_error(y_test, y_pred)
initial_mse
```

#### средняя абсолютная ошибка

```python
(np.abs(check_test['error'])).mean()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

#### R**2

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

#### просмотр коэффициентов линейной регрессии

#### W0

```python
lr.intercept_
```

#### W1...Wm

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.barh(x_trane.columns, lr.coef_.flatten())
plt.xlabel('вес признака')
plt.ylabel('признак')
```

#### Стандартизация признаков

При стандартизации от признака надо отнять среднее и поделить на среднеквадратичное отклонение

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_trane_scaled = scaler.fit_transform(x_trane)
```

```python
x_trane_scaled = pd.DataFrame(x_trane_scaled, columns=x_trane.columns)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
lr.fit(x_trane_scaled, y_trane)
```
